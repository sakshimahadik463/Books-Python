```python

import matplotlib.pyplot as plt 
import numpy as np
```

```python
##you can use the axis parameter in the grid() functions to specify which lines to display.

## legal values are: 'x,'y',and 'both'. default value is 'both 

x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260 ,270 ,280, 290,300,310,320,330])

plt.title('sports watch data')
plt.xlabel('average pulse')
plt.ylabel('calprie burnage')

plt.plot(x,y,color='orange')
plt.grid(axis='x')
plt.show()
```

```python
## set line propertiese for the grid 
## you can also set the line propertiese of the grid , like this: grid
## (color = 'color' linestyle='linestyle,linewidth=number).
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260 ,270 ,280, 290,300,310,320,330])

plt.title('sports watch data')
plt.xlabel('average pulse')
plt.ylabel('calprie burnage')

plt.plot(x,y,color='orange')
plt.grid(color='hotpink',linestyle='--',linewidth=1.5)
plt.show()
```
