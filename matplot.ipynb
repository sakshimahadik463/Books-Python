```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
### Draw a line in a diagram from position (0,10) to position (5,150):
xpoints=np.array([1,2])
ypoints=np.array([5,10])

plt.plot(xpoints,ypoints)
plt.show()
```

```python
x=np.array([1,20])
y=np.array([10,200])

plt.plot(x,y)
plt.show
```

```python
x=np.array([1,6])
y=np.array([4,5])
plt.plot(x,y,'+')
plt.show()
```

```python
y=np.array([3,8,1,12])
plt.plot(y,'*''r')
plt.show
```

```python
##plt.plot(x,y)    we can specify symbol and colour too '*''r'
```

```python
y=np.array([3,8,1,12])
x=np.array([5,6,7,12])
plt.plot(x,y,marker='*',ms=50,color='pink')  ## if we want to plot the points connected with lines then we will introduce marker 
plt.show()
```

```python
y=np.array([3,8,1,12])
x=np.array([5,6,7,12])
plt.plot(x,y,'*''r')    ##if we want to plot points distributed write '*''r'    
plt.show()
```

```python
   y=np.array([3,8,1])
   plt.plot(y,marker='o',ms=50,mec='b',color='pink')    ##mec = marker edge color
   plt.show()
```

```python
y=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',ms=20,mfc='r',mec='k')   ##mfc = marker face color
plt.show()
```

```python
### use both mec and mfc arguments to color of the entire marker  
y=np.array([1,3,5,8])
plt.plot(y,marker='o',ms=20,mec='b',mfc='k')
plt.show()
```

```python
y=np.array([3,8,1,10])
plt.plot(y,linestyle='dashed',marker='o',mec='hotpink',ms=20,mfc='blue',linewidth=10)
plt.show()
```
